```python
#Importing Libraries and Creating a Spark Session
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

spark = SparkSession.builder.getOrCreate()

print("success")
```

```python
fd_path = "C:/Users/tf/Desktop/airbnb_data_cleaned.csv"

df = spark.read.csv(fd_path, header=True, inferSchema=True)

total_rows = df.count()  
total_columns = len(df.columns)

print("Dataset loaded successfully")
print("Number of rows:", total_rows, "Number of columns:", total_columns)
```
